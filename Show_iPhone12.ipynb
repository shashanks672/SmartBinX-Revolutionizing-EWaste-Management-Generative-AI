```python
# Step 2: Read the iPhone 12 data from the database
import sqlite3, json

# open your database file
conn = sqlite3.connect("smartbinx_full.db")
cur = conn.cursor()

# read the latest entry
cur.execute("SELECT display_name, source, materials_json FROM ewaste_models ORDER BY last_updated DESC LIMIT 1")
row = cur.fetchone()
conn.close()

# show it neatly
print("Device name:", row[0])
print("Source:", row[1])
materials = json.loads(row[2])
print("\n--- Material Composition ---")
for k,v in materials.items():
    print(f"{k:12}: {v}%")
```
